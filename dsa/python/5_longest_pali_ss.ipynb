```python
from typing import NamedTuple


class Palindrom(NamedTuple):
    length: int
    start: int
    end: int


def isPali(st: str) -> bool:
    return st == st[::-1]


def longestPali(
    s: str,
    debug_print: bool = True,
    existing: None | list[Palindrom] = None,
    it: int = 0,
) -> str:
    if isPali(s):
        return s
    if len(s) < 2:
        return s
    elif len(s) == 2:
        return s if s[0] == s[1] else s[0]

    if existing:
        candidates = existing
    else:
        candidates: list[Palindrom] = [Palindrom(1, i, i) for i, _ in enumerate(s)]

    uneven_length_solutions: set[Palindrom] = set()
    hall_of_fame: set[Palindrom] = set()

    for p in candidates:
        n, i, j = p
        while i > 0 and j < len(s) - 1:
            if s[i - 1] == s[j + 1]:
                n += 2
                i -= 1
                j += 1
            else:
                break

        uneven_length_solutions.add(Palindrom(n, i, j))

    for length, i, j in uneven_length_solutions:
        while i > 0:
            if isPali(s[i - 1 : j + 1]):
                i -= 1
                length += 1
                continue
            else:
                break
        while j < len(s) - 1:
            if isPali(s[i : j + 2]):
                j += 1
                length += 1
                continue
            else:
                break
        hall_of_fame.add(Palindrom(length, i, j))

    solutions = [p for p in hall_of_fame]
    solutions.sort(key=lambda x: x.length, reverse=True)

    if debug_print:
        for l, mm, nn in hall_of_fame:
            print(l, s[mm : nn + 1], "even" if l % 2 == 0 else "uneven")

    it += 1
    if it == 2:
        _, m, n = solutions[0]

        return s[m : n + 1]
    else:
        return longestPali(s, debug_print, solutions, it)


class Solution:
    def longestPalindrome(self, s: str) -> str:
        return longestPali(s, debug_print=False)
```
