```python
def f(x: int, min_val: int = -(2**31), max_val: int = 2**31 - 1) -> int:
    sign = 1 if x >= 0 else -1
    match sign:
        case 1:
            rev_str = "".join((reversed(str(x))))
        case -1:
            rev_str = "".join((reversed(str(x)[1:])))

    rev_int = int(rev_str)
    if min_val <= rev_int <= max_val:
        return sign * rev_int
    return 0


class Solution:
    def reverse(self, x: int) -> int:
        return f(x)
```
